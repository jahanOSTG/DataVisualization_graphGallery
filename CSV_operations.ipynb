### iloc means "integer location" â€“ it helps you select rows and columns in a DataFrame by number, not by name.
- DataFrame.iloc[row_index, column_index]

```python
import matplotlib.pyplot as plt

# Line plot of students' scores (first 10 students)
show.iloc[:10].plot(x='Name', y=['Math', 'English', 'Science'], kind='line')
plt.title("Scores of First 10 Students")
plt.ylabel("Score")
plt.show()
```

### Bar chart of English marks 

```python
show.iloc[:5].plot(x='Name',y='English',kind='bar',color='pink')
plt.title("English Rate")
plt.ylabel("English Marks")
plt.show()
```

### Histogram of Mathematics scores

```python
show['Science'].plot(kind='hist', bins=10, color='cyan')
plt.title("Distribution of Science Scores")
plt.xlabel("Score")
plt.show()
```

### Scatter plot: Mathematics vs Science

```python
show.plot.scatter(x='Math', y='Science', color='pink')
plt.title("Mathematics vs Science Scores")
plt.show()
```

## Box plot for all subjects

```python

show[['Math', 'English', 'Science']].plot.box()
plt.title("Score Distribution Boxplot")
plt.show()
```
